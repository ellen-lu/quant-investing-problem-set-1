Calculate the sample mean and standard deviation of returns for each of the four equal-weight portfolios. Plot estimated standard deviations as a function of the number of stocks in the equal-weight portfolio.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#import dataset
df =pd.read_excel(r'C:\A++Yale\A+AM\Academics\MGT 595\Problem Set\PSet 1\Problem_Set1.xlsx',
                  parse_dates=["date"],index_col="date",skiprows=5)
df.head()
```

```python
#Equal-weight portfolio using the first 5 stocks
data1=df.iloc[:,:5]
data1.head()

#Sample Mean
df_output1 = pd.DataFrame(index = data1.mean().index, data = data1.mean().values, columns = ['Sample_Mean'])
variances = data1.var(axis=0)
num_assets = len(data1.columns)
weights = [1 / num_assets for i in range (num_assets)]
weights
df_output1['Sample_Variance']=variances
df_output1['Weights']=weights
df_output1

ptf_mean5 = sum(df_output1["Sample_Mean"]*df_output1["Weights"])

#Sample Standard Deviation
stock_cov = np.array(data1.cov())
w = np.array(np.asmatrix(weights))

ptf_var = np.dot(w, np.dot(stock_cov,w.T))
ptf_std5 = np.sqrt(ptf_var)

print("Sample mean return of a portfolio of first 5 stocks is: "+ str(ptf_mean5)+ " percent")
print("Sample standard deviation of a portfolio of first 5 stocks is: "+ str(ptf_std5))
```
